```python
# CELL 21: Create data splits as per spec (Section 3.4)

import json
import random

random.seed(42)  # Reproducibility

print("📊 Creating data splits...")
print("=" * 60)

# Splits per level (from spec)
TRAIN_SIZE = 20000
TEST_SIZE = 1000
ANALYSIS_SIZE = 100

data_splits = {}
split_indices = {}

for level, ds in datasets_raw.items():
    total = len(ds)
    
    # Create random indices
    all_indices = list(range(total))
    random.shuffle(all_indices)
    
    # Split indices
    train_idx = all_indices[:TRAIN_SIZE]
    test_idx = all_indices[TRAIN_SIZE:TRAIN_SIZE + TEST_SIZE]
    analysis_idx = all_indices[TRAIN_SIZE + TEST_SIZE:TRAIN_SIZE + TEST_SIZE + ANALYSIS_SIZE]
    
    # Store indices for reproducibility
    split_indices[level] = {
        "train": train_idx,
        "test": test_idx,
        "analysis": analysis_idx
    }
    
    # Create actual splits
    data_splits[level] = {
        "train": ds.select(train_idx),
        "test": ds.select(test_idx),
        "analysis": ds.select(analysis_idx)
    }
    
    print(f"  {level}: train={len(train_idx)}, test={len(test_idx)}, analysis={len(analysis_idx)}")

# Save split indices to JSON (for reproducibility)
with open("/workspace/data/data_splits.json", "w") as f:
    json.dump(split_indices, f)

print("\n✅ Saved split indices to /workspace/data/data_splits.json")

# Summary
print("\n" + "=" * 60)
print("📋 SUMMARY:")
print(f"   Total training: {TRAIN_SIZE * 5:,} samples")
print(f"   Total test: {TEST_SIZE * 5:,} samples")
print(f"   Total analysis: {ANALYSIS_SIZE * 5:,} samples")
print("=" * 60)
```

```python
# CELL 22: Create formatted prompts and save data files

import json

print("📝 Formatting prompts and saving data...")
print("=" * 60)

def format_prompt(example):
    """Format a single example into the prompt template from spec Section 3.5"""
    prompt = f"""You are a SQL expert. Your role is to generate SQL statements.

Given the following table schema:
{example['create_statement']}

Generate a SQL statement for: "{example['english_prompt']}"

SQL:"""
    return prompt

def format_dataset(ds):
    """Format entire dataset with prompts"""
    formatted = []
    for example in ds:
        formatted.append({
            "prompt": format_prompt(example),
            "completion": example['sql_statement'],
            "english": example['english_prompt'],
            "schema": example['create_statement']
        })
    return formatted

# Process and save each level
for level in ["CS1", "CS2", "CS3", "CS4", "CS5"]:
    print(f"\n  Processing {level}...")
    
    # Format train
    train_formatted = format_dataset(data_splits[level]["train"])
    with open(f"/workspace/data/train_{level}.json", "w") as f:
        json.dump(train_formatted, f)
    print(f"    ✅ train_{level}.json ({len(train_formatted)} samples)")
    
    # Format test  
    test_formatted = format_dataset(data_splits[level]["test"])
    with open(f"/workspace/data/test_{level}.json", "w") as f:
        json.dump(test_formatted, f)
    print(f"    ✅ test_{level}.json ({len(test_formatted)} samples)")
    
    # Format analysis
    analysis_formatted = format_dataset(data_splits[level]["analysis"])
    with open(f"/workspace/data/analysis_{level}.json", "w") as f:
        json.dump(analysis_formatted, f)
    print(f"    ✅ analysis_{level}.json ({len(analysis_formatted)} samples)")

print("\n" + "=" * 60)
print("📋 Let's see one formatted example:")
print("=" * 60)

# Show one example
with open("/workspace/data/train_CS1.json", "r") as f:
    sample = json.load(f)[0]

print("\n🔹 PROMPT:")
print(sample["prompt"][:500] + "..." if len(sample["prompt"]) > 500 else sample["prompt"])
print("\n🔹 EXPECTED COMPLETION:")
print(sample["completion"])
```

```python
# CELL 30: Measure base model accuracy (pre fine-tuning baseline)

import json
from tqdm.auto import tqdm

print("📊 Measuring BASE model accuracy (before fine-tuning)...")
print("=" * 60)

def evaluate_model(model, tokenizer, test_data, n_samples=50):
    """Evaluate model on test samples"""
    correct = 0
    total = min(n_samples, len(test_data))
    
    for i in tqdm(range(total), desc="Evaluating"):
        example = test_data[i]
        prompt = example["prompt"]
        expected = example["completion"].strip()
        
        # Generate
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=150,
                do_sample=False,
                pad_token_id=tokenizer.eos_token_id
            )
        
        generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_sql = generated[len(prompt):].strip().split("\n")[0].strip()
        
        # Exact match
        if generated_sql == expected:
            correct += 1
    
    return correct / total

# Test on 50 samples per level (quick baseline)
base_accuracy = {}

for level in ["CS1", "CS2", "CS3", "CS4", "CS5"]:
    print(f"\n🔹 Testing {level}...")
    with open(f"/workspace/data/test_{level}.json", "r") as f:
        test_data = json.load(f)
    
    acc = evaluate_model(model, tokenizer, test_data, n_samples=50)
    base_accuracy[level] = acc
    print(f"   Accuracy: {acc*100:.1f}%")

# Save baseline
baseline_results = {
    "model": "base",
    "samples_per_level": 50,
    "accuracy": base_accuracy
}

with open("/workspace/results/baseline/accuracy_base_model.json", "w") as f:
    json.dump(baseline_results, f, indent=2)

print("\n" + "=" * 60)
print("📋 BASE MODEL ACCURACY:")
for level, acc in base_accuracy.items():
    print(f"   {level}: {acc*100:.1f}%")
print(f"\n   Average: {sum(base_accuracy.values())/5*100:.1f}%")
print("=" * 60)
print("\n✅ Saved to /workspace/results/baseline/accuracy_base_model.json")
```

```python
# CELL 31: Prepare for Fine-Tuning

import json
from torch.utils.data import Dataset, DataLoader

print("🔧 Preparing for fine-tuning...")
print("=" * 60)

# Custom Dataset class
class SQLDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=512):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        item = self.data[idx]
        
        # Combine prompt + completion
        full_text = item["prompt"] + item["completion"] + self.tokenizer.eos_token
        
        # Tokenize
        encoded = self.tokenizer(
            full_text,
            truncation=True,
            max_length=self.max_length,
            padding="max_length",
            return_tensors="pt"
        )
        
        input_ids = encoded["input_ids"].squeeze()
        attention_mask = encoded["attention_mask"].squeeze()
        
        # Labels = input_ids (for causal LM)
        labels = input_ids.clone()
        
        # Mask padding tokens in labels
        labels[labels == self.tokenizer.pad_token_id] = -100
        
        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": labels
        }

# Load ALL training data (combined)
print("Loading training data...")
all_train_data = []
for level in ["CS1", "CS2", "CS3", "CS4", "CS5"]:
    with open(f"/workspace/data/train_{level}.json", "r") as f:
        data = json.load(f)
        all_train_data.extend(data)
        print(f"  {level}: {len(data)} samples")

print(f"\n  Total: {len(all_train_data)} samples")

# Set pad token
tokenizer.pad_token = tokenizer.eos_token

# Create dataset
train_dataset = SQLDataset(all_train_data, tokenizer)

print(f"\n✅ Dataset ready: {len(train_dataset)} samples")
print("=" * 60)
```

```python
# CELL 2: Test accuracy on all complexity levels

import json
from tqdm.auto import tqdm

print("📊 TESTING FINE-TUNED MODEL ACCURACY")
print("=" * 60)

def test_accuracy(model, tokenizer, test_file, n_samples=50):
    """Test model on samples and return accuracy"""
    with open(test_file, "r") as f:
        test_data = json.load(f)
    
    correct = 0
    total = min(n_samples, len(test_data))
    
    for i in tqdm(range(total), desc="Testing", leave=False):
        example = test_data[i]
        prompt = example["prompt"]
        expected = example["completion"].strip()
        
        # Generate
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=150,
                do_sample=False,
                pad_token_id=tokenizer.eos_token_id
            )
        
        generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_sql = generated[len(prompt):].strip().split("\n")[0].strip()
        
        if generated_sql == expected:
            correct += 1
    
    return correct / total

# Test on each complexity level
results = {}
for level in ["CS1", "CS2", "CS3", "CS4", "CS5"]:
    print(f"\n🔹 Testing {level}...")
    acc = test_accuracy(finetuned_model, tokenizer, f"/workspace/data/test_{level}.json", n_samples=50)
    results[level] = acc
    print(f"   Accuracy: {acc*100:.1f}%")

# Summary
print("\n" + "=" * 60)
print("📋 FINE-TUNED MODEL ACCURACY:")
print("=" * 60)
for level, acc in results.items():
    bar = "█" * int(acc * 20) + "░" * (20 - int(acc * 20))
    print(f"   {level}: {bar} {acc*100:.1f}%")

avg_acc = sum(results.values()) / len(results)
print(f"\n   AVERAGE: {avg_acc*100:.1f}%")
print("=" * 60)

# Save results
finetuned_results = {
    "model": "finetuned",
    "checkpoint": "checkpoint-5500",
    "samples_per_level": 50,
    "accuracy": results,
    "average_accuracy": avg_acc
}

with open("/workspace/results/baseline/accuracy_finetuned_model.json", "w") as f:
    json.dump(finetuned_results, f, indent=2)

print("✅ Saved to /workspace/results/baseline/accuracy_finetuned_model.json")
```

```python
# CELL 4: Comprehensive Weight Deltas Analysis

import torch
import json
from collections import defaultdict

print("📊 COMPREHENSIVE WEIGHT DELTAS ANALYSIS")
print("=" * 60)

# 1. Load base model
print("1️⃣ Loading base model...")
base_model = AutoModelForCausalLM.from_pretrained(
    "/workspace/models/base",
    torch_dtype=torch.bfloat16,
    device_map="cpu"
)
print("   ✅ Base model loaded")

# 2. Compute all weight deltas
print("\n2️⃣ Computing weight deltas for all parameters...")

weight_deltas = {}
base_params = dict(base_model.named_parameters())

for name, param in finetuned_model.named_parameters():
    if name in base_params:
        base_param = base_params[name]
        
        # Compute delta
        delta = (param.data.cpu().float() - base_param.data.cpu().float())
        delta_norm = delta.norm().item()
        base_norm = base_param.data.cpu().float().norm().item()
        relative_change = delta_norm / (base_norm + 1e-8)
        
        weight_deltas[name] = {
            "delta_norm": delta_norm,
            "base_norm": base_norm,
            "relative_change": relative_change,
            "shape": list(param.shape),
            "num_params": param.numel()
        }

print(f"   ✅ Computed deltas for {len(weight_deltas)} parameters")

# 3. Analyze by LAYER NUMBER
print("\n" + "=" * 60)
print("📈 CHANGE BY LAYER NUMBER")
print("=" * 60)

layer_changes = defaultdict(lambda: {"attention": 0, "mlp": 0, "norm": 0, "count": {"attention": 0, "mlp": 0, "norm": 0}})

for name, stats in weight_deltas.items():
    # Extract layer number
    if "layers." in name:
        layer_num = int(name.split("layers.")[1].split(".")[0])
        
        if "self_attn" in name:
            layer_changes[layer_num]["attention"] += stats["relative_change"]
            layer_changes[layer_num]["count"]["attention"] += 1
        elif "mlp" in name:
            layer_changes[layer_num]["mlp"] += stats["relative_change"]
            layer_changes[layer_num]["count"]["mlp"] += 1
        elif "norm" in name:
            layer_changes[layer_num]["norm"] += stats["relative_change"]
            layer_changes[layer_num]["count"]["norm"] += 1

# Print per-layer changes
print(f"\n{'Layer':<8} {'Attention':>12} {'MLP':>12} {'Total':>12}")
print("-" * 48)

layer_totals = []
for layer_num in sorted(layer_changes.keys()):
    data = layer_changes[layer_num]
    attn_avg = data["attention"] / max(data["count"]["attention"], 1)
    mlp_avg = data["mlp"] / max(data["count"]["mlp"], 1)
    total_avg = (attn_avg + mlp_avg) / 2
    layer_totals.append((layer_num, attn_avg, mlp_avg, total_avg))
    
    attn_bar = "█" * int(attn_avg * 300)
    mlp_bar = "█" * int(mlp_avg * 300)
    print(f"L{layer_num:<7} {attn_avg*100:>10.3f}% {mlp_avg*100:>10.3f}% {total_avg*100:>10.3f}%")

# 4. Analyze by COMPONENT TYPE (detailed)
print("\n" + "=" * 60)
print("🔧 CHANGE BY COMPONENT TYPE (Detailed)")
print("=" * 60)

component_stats = defaultdict(lambda: {"total_delta": 0, "count": 0, "params": []})

for name, stats in weight_deltas.items():
    # Categorize
    if "q_proj" in name:
        comp = "attention.Q"
    elif "k_proj" in name:
        comp = "attention.K"
    elif "v_proj" in name:
        comp = "attention.V"
    elif "o_proj" in name:
        comp = "attention.O"
    elif "gate_proj" in name:
        comp = "mlp.gate"
    elif "up_proj" in name:
        comp = "mlp.up"
    elif "down_proj" in name:
        comp = "mlp.down"
    elif "input_layernorm" in name:
        comp = "layernorm.input"
    elif "post_attention_layernorm" in name:
        comp = "layernorm.post_attn"
    elif "embed_tokens" in name:
        comp = "embedding"
    elif "lm_head" in name:
        comp = "lm_head"
    elif "norm" in name:
        comp = "final_norm"
    else:
        comp = "other"
    
    component_stats[comp]["total_delta"] += stats["relative_change"]
    component_stats[comp]["count"] += 1

print(f"\n{'Component':<25} {'Avg Change':>12} {'Visual':>25}")
print("-" * 65)

sorted_components = sorted(component_stats.items(), key=lambda x: x[1]["total_delta"]/max(x[1]["count"],1), reverse=True)

for comp, data in sorted_components:
    avg_change = data["total_delta"] / max(data["count"], 1)
    bar = "█" * int(avg_change * 400) + "░" * max(0, 20 - int(avg_change * 400))
    print(f"{comp:<25} {avg_change*100:>10.4f}% {bar}")

# 5. Find TOP 20 most changed individual weights
print("\n" + "=" * 60)
print("🔝 TOP 20 MOST CHANGED PARAMETERS")
print("=" * 60)

sorted_all = sorted(weight_deltas.items(), key=lambda x: x[1]["relative_change"], reverse=True)

for i, (name, stats) in enumerate(sorted_all[:20]):
    short_name = name.replace("model.", "").replace("layers.", "L")
    print(f"{i+1:>2}. {short_name[:55]:<55} Δ={stats['relative_change']*100:.3f}%")

# 6. Visualize layer heatmap (text-based)
print("\n" + "=" * 60)
print("🗺️ LAYER CHANGE HEATMAP")
print("=" * 60)

max_change = max([t[3] for t in layer_totals]) if layer_totals else 1

print("\nAttention changes by layer:")
attn_line = ""
for layer_num, attn, mlp, total in layer_totals:
    intensity = int(attn / (max([t[1] for t in layer_totals]) + 1e-8) * 9)
    chars = " ░▒▓█▓▒░ "[max(0, min(8, intensity))]
    attn_line += chars
print(f"[{attn_line}]")
print(f" {'Early':<15}{'Middle':<15}{'Late':>10}")

print("\nMLP changes by layer:")
mlp_line = ""
for layer_num, attn, mlp, total in layer_totals:
    intensity = int(mlp / (max([t[2] for t in layer_totals]) + 1e-8) * 9)
    chars = " ░▒▓█▓▒░ "[max(0, min(8, intensity))]
    mlp_line += chars
print(f"[{mlp_line}]")
print(f" {'Early':<15}{'Middle':<15}{'Late':>10}")

# 7. Save comprehensive results
print("\n" + "=" * 60)
print("💾 SAVING RESULTS...")

# Full weight deltas
with open("/workspace/results/weights/weight_deltas_full.json", "w") as f:
    json.dump(weight_deltas, f, indent=2)

# Per-layer summary
layer_summary = {f"layer_{l}": {"attention": a*100, "mlp": m*100, "total": t*100} 
                 for l, a, m, t in layer_totals}
with open("/workspace/results/weights/per_layer_changes.json", "w") as f:
    json.dump(layer_summary, f, indent=2)

# Component summary
component_summary = {comp: {"avg_change_percent": data["total_delta"]/max(data["count"],1)*100, "count": data["count"]} 
                     for comp, data in component_stats.items()}
with open("/workspace/results/weights/component_changes.json", "w") as f:
    json.dump(component_summary, f, indent=2)

print("   ✅ /workspace/results/weights/weight_deltas_full.json")
print("   ✅ /workspace/results/weights/per_layer_changes.json")
print("   ✅ /workspace/results/weights/component_changes.json")

# Cleanup
del base_model
torch.cuda.empty_cache()

print("\n" + "=" * 60)
print("✅ WEIGHT DELTA ANALYSIS COMPLETE!")
print("=" * 60)
```

```python
# CELL 6: Detailed Architecture & Weight-Based Neuron Analysis

import torch
import json
import numpy as np

print("🔬 DETAILED ARCHITECTURE & NEURON ANALYSIS")
print("=" * 60)

# 1. Model Architecture Details
print("\n1️⃣ GEMMA 2 2B ARCHITECTURE:")
print("-" * 50)

config = finetuned_model.config
print(f"   Layers: {config.num_hidden_layers}")
print(f"   Hidden size: {config.hidden_size}")
print(f"   Intermediate (MLP) size: {config.intermediate_size}")
print(f"   Attention heads: {config.num_attention_heads}")
print(f"   Head dimension: {config.head_dim}")
print(f"   Vocab size: {config.vocab_size}")

# 2. Load base model for comparison
print("\n2️⃣ Loading base model for neuron analysis...")
base_model = AutoModelForCausalLM.from_pretrained(
    "/workspace/models/base",
    torch_dtype=torch.bfloat16,
    device_map="cpu"
)

# 3. Analyze MLP neurons (most changed component)
print("\n3️⃣ MLP NEURON ANALYSIS (per layer):")
print("-" * 50)
print(f"   Each layer has {config.intermediate_size} MLP neurons")
print(f"   Analyzing which neurons changed most...\n")

neuron_analysis = {}

for layer_idx in range(config.num_hidden_layers):
    # Get gate_proj weights (determines which neurons activate)
    ft_gate = finetuned_model.model.layers[layer_idx].mlp.gate_proj.weight.data.cpu().float()
    base_gate = base_model.model.layers[layer_idx].mlp.gate_proj.weight.data.cpu().float()
    
    # Get down_proj weights (most changed - output of MLP)
    ft_down = finetuned_model.model.layers[layer_idx].mlp.down_proj.weight.data.cpu().float()
    base_down = base_model.model.layers[layer_idx].mlp.down_proj.weight.data.cpu().float()
    
    # Per-neuron change (L2 norm of weight change for each neuron)
    # down_proj shape: [hidden_size, intermediate_size] -> each column is a neuron
    neuron_deltas = (ft_down - base_down).norm(dim=0)  # [intermediate_size]
    
    # Neuron importance (by weight magnitude)
    neuron_importance_ft = ft_down.norm(dim=0)
    neuron_importance_base = base_down.norm(dim=0)
    
    # Statistics
    top_changed_neurons = neuron_deltas.topk(10).indices.tolist()
    top_changed_values = neuron_deltas.topk(10).values.tolist()
    
    # Near-zero neurons (potential "dead" by weight)
    threshold = 0.01
    near_zero_base = (neuron_importance_base < threshold).sum().item()
    near_zero_ft = (neuron_importance_ft < threshold).sum().item()
    
    neuron_analysis[f"layer_{layer_idx}"] = {
        "total_neurons": config.intermediate_size,
        "avg_delta": neuron_deltas.mean().item(),
        "max_delta": neuron_deltas.max().item(),
        "top_10_changed_neurons": top_changed_neurons,
        "top_10_delta_values": top_changed_values,
        "near_zero_base": near_zero_base,
        "near_zero_finetuned": near_zero_ft
    }
    
    if layer_idx % 5 == 0 or layer_idx == 25:  # Print every 5th layer
        print(f"   L{layer_idx}: avg_Δ={neuron_deltas.mean().item():.4f}, "
              f"max_Δ={neuron_deltas.max().item():.4f}, "
              f"top_neuron={top_changed_neurons[0]}")

# 4. Find globally most changed neurons
print("\n4️⃣ TOP 20 MOST CHANGED MLP NEURONS (globally):")
print("-" * 50)

all_neurons = []
for layer_idx in range(config.num_hidden_layers):
    ft_down = finetuned_model.model.layers[layer_idx].mlp.down_proj.weight.data.cpu().float()
    base_down = base_model.model.layers[layer_idx].mlp.down_proj.weight.data.cpu().float()
    neuron_deltas = (ft_down - base_down).norm(dim=0)
    
    for neuron_idx, delta in enumerate(neuron_deltas):
        all_neurons.append((layer_idx, neuron_idx, delta.item()))

all_neurons.sort(key=lambda x: x[2], reverse=True)

top_20_neurons = []
for i, (layer, neuron, delta) in enumerate(all_neurons[:20]):
    print(f"   {i+1:>2}. Layer {layer:>2}, Neuron {neuron:>4}: Δ = {delta:.4f}")
    top_20_neurons.append({"layer": layer, "neuron": neuron, "delta": delta})

# 5. Attention head analysis
print("\n5️⃣ ATTENTION HEAD ANALYSIS:")
print("-" * 50)

head_analysis = {}
for layer_idx in range(config.num_hidden_layers):
    # Q projection change
    ft_q = finetuned_model.model.layers[layer_idx].self_attn.q_proj.weight.data.cpu().float()
    base_q = base_model.model.layers[layer_idx].self_attn.q_proj.weight.data.cpu().float()
    
    # Reshape to [num_heads, head_dim, hidden_size]
    num_heads = config.num_attention_heads
    head_dim = config.head_dim
    
    ft_q_heads = ft_q.view(num_heads, head_dim, -1)
    base_q_heads = base_q.view(num_heads, head_dim, -1)
    
    # Per-head change
    head_deltas = (ft_q_heads - base_q_heads).norm(dim=(1,2))  # [num_heads]
    
    head_analysis[f"layer_{layer_idx}"] = {
        "head_deltas": head_deltas.tolist(),
        "most_changed_head": head_deltas.argmax().item(),
        "max_delta": head_deltas.max().item()
    }
    
    if layer_idx % 5 == 0 or layer_idx == 25:
        print(f"   L{layer_idx}: most_changed_head={head_deltas.argmax().item()}, "
              f"delta={head_deltas.max().item():.4f}")

# 6. Save all analysis
print("\n6️⃣ SAVING DETAILED ANALYSIS...")

# Save neuron analysis
with open("/workspace/results/weights/neuron_analysis.json", "w") as f:
    json.dump(neuron_analysis, f, indent=2)

# Save top neurons
with open("/workspace/results/weights/top_changed_neurons.json", "w") as f:
    json.dump(top_20_neurons, f, indent=2)

# Save head analysis
with open("/workspace/results/weights/attention_head_analysis.json", "w") as f:
    json.dump(head_analysis, f, indent=2)

print("   ✅ /workspace/results/weights/neuron_analysis.json")
print("   ✅ /workspace/results/weights/top_changed_neurons.json")
print("   ✅ /workspace/results/weights/attention_head_analysis.json")

# Cleanup
del base_model
torch.cuda.empty_cache()

print("\n" + "=" * 60)
print("✅ DETAILED NEURON ANALYSIS COMPLETE!")
print("=" * 60)

# Summary
print("\n📊 KEY FINDINGS:")
print(f"   • Total MLP neurons per layer: {config.intermediate_size}")
print(f"   • Most changed neuron: Layer {top_20_neurons[0]['layer']}, "
      f"Neuron {top_20_neurons[0]['neuron']} (Δ={top_20_neurons[0]['delta']:.4f})")
print(f"   • SQL circuits likely concentrated in top changed neurons")
```

```python
# CELL 6: PRECISE Architecture & Full Component Analysis

import torch
import json
import numpy as np
from collections import defaultdict

print("🔬 PRECISE ARCHITECTURE & FULL COMPONENT ANALYSIS")
print("=" * 70)

# 1. EXACT Model Architecture
print("\n" + "=" * 70)
print("1️⃣ EXACT GEMMA 2 2B ARCHITECTURE")
print("=" * 70)

config = finetuned_model.config

architecture = {
    "model_name": "google/gemma-2-2b",
    "num_layers": config.num_hidden_layers,
    "hidden_size": config.hidden_size,
    "intermediate_size": config.intermediate_size,
    "num_attention_heads": config.num_attention_heads,
    "num_key_value_heads": config.num_key_value_heads,
    "head_dim": config.head_dim,
    "vocab_size": config.vocab_size,
    "max_position_embeddings": config.max_position_embeddings,
    "rms_norm_eps": config.rms_norm_eps,
    "attention_type": "Grouped Query Attention (GQA)",
}

print(f"""
┌─────────────────────────────────────────────────────────────────┐
│                    GEMMA 2 2B ARCHITECTURE                      │
├─────────────────────────────────────────────────────────────────┤
│  Total Layers:          {config.num_hidden_layers:>5}                                  │
│  Hidden Dimension:      {config.hidden_size:>5}                                  │
│  MLP Dimension:         {config.intermediate_size:>5}  (4x hidden)                     │
│  Vocab Size:            {config.vocab_size:>6}                                 │
├─────────────────────────────────────────────────────────────────┤
│  ATTENTION (Grouped Query Attention):                           │
│    • Query Heads:       {config.num_attention_heads:>5}  per layer                       │
│    • Key/Value Heads:   {config.num_key_value_heads:>5}  per layer (shared)              │
│    • Head Dimension:    {config.head_dim:>5}                                  │
│    • Q proj shape:      [{config.num_attention_heads * config.head_dim}, {config.hidden_size}]                       │
│    • K proj shape:      [{config.num_key_value_heads * config.head_dim}, {config.hidden_size}]                       │
│    • V proj shape:      [{config.num_key_value_heads * config.head_dim}, {config.hidden_size}]                       │
│    • O proj shape:      [{config.hidden_size}, {config.num_attention_heads * config.head_dim}]                       │
├─────────────────────────────────────────────────────────────────┤
│  MLP (SwiGLU):                                                  │
│    • gate_proj:         [{config.intermediate_size}, {config.hidden_size}]                       │
│    • up_proj:           [{config.intermediate_size}, {config.hidden_size}]                       │
│    • down_proj:         [{config.hidden_size}, {config.intermediate_size}]                       │
│    • Neurons per layer: {config.intermediate_size:>5}                                  │
│    • Total MLP neurons: {config.intermediate_size * config.num_hidden_layers:>6}                                 │
└─────────────────────────────────────────────────────────────────┘
""")

# 2. Load base model
print("2️⃣ Loading base model for precise comparison...")
base_model = AutoModelForCausalLM.from_pretrained(
    "/workspace/models/base",
    torch_dtype=torch.bfloat16,
    device_map="cpu"
)
print("   ✅ Base model loaded")

# 3. PRECISE ATTENTION HEAD ANALYSIS (ALL HEADS)
print("\n" + "=" * 70)
print("3️⃣ ATTENTION HEAD ANALYSIS (ALL 8 HEADS × 26 LAYERS = 208 HEADS)")
print("=" * 70)

attention_head_data = {}

for layer_idx in range(config.num_hidden_layers):
    layer_data = {
        "layer": layer_idx,
        "q_heads": [],
        "k_heads": [],
        "v_heads": [],
        "o_heads": [],
    }
    
    # Q projection: [num_heads * head_dim, hidden_size]
    ft_q = finetuned_model.model.layers[layer_idx].self_attn.q_proj.weight.data.cpu().float()
    base_q = base_model.model.layers[layer_idx].self_attn.q_proj.weight.data.cpu().float()
    
    # K projection: [num_kv_heads * head_dim, hidden_size]
    ft_k = finetuned_model.model.layers[layer_idx].self_attn.k_proj.weight.data.cpu().float()
    base_k = base_model.model.layers[layer_idx].self_attn.k_proj.weight.data.cpu().float()
    
    # V projection: [num_kv_heads * head_dim, hidden_size]
    ft_v = finetuned_model.model.layers[layer_idx].self_attn.v_proj.weight.data.cpu().float()
    base_v = base_model.model.layers[layer_idx].self_attn.v_proj.weight.data.cpu().float()
    
    # O projection: [hidden_size, num_heads * head_dim]
    ft_o = finetuned_model.model.layers[layer_idx].self_attn.o_proj.weight.data.cpu().float()
    base_o = base_model.model.layers[layer_idx].self_attn.o_proj.weight.data.cpu().float()
    
    # Analyze Q heads (8 heads)
    for head_idx in range(config.num_attention_heads):
        start = head_idx * config.head_dim
        end = (head_idx + 1) * config.head_dim
        
        q_delta = (ft_q[start:end, :] - base_q[start:end, :]).norm().item()
        q_base_norm = base_q[start:end, :].norm().item()
        q_ft_norm = ft_q[start:end, :].norm().item()
        
        layer_data["q_heads"].append({
            "head_idx": head_idx,
            "delta_norm": q_delta,
            "base_norm": q_base_norm,
            "ft_norm": q_ft_norm,
            "relative_change": q_delta / (q_base_norm + 1e-8)
        })
    
    # Analyze K heads (4 heads - GQA)
    for head_idx in range(config.num_key_value_heads):
        start = head_idx * config.head_dim
        end = (head_idx + 1) * config.head_dim
        
        k_delta = (ft_k[start:end, :] - base_k[start:end, :]).norm().item()
        k_base_norm = base_k[start:end, :].norm().item()
        
        layer_data["k_heads"].append({
            "head_idx": head_idx,
            "delta_norm": k_delta,
            "base_norm": k_base_norm,
            "relative_change": k_delta / (k_base_norm + 1e-8)
        })
    
    # Analyze V heads (4 heads - GQA)
    for head_idx in range(config.num_key_value_heads):
        start = head_idx * config.head_dim
        end = (head_idx + 1) * config.head_dim
        
        v_delta = (ft_v[start:end, :] - base_v[start:end, :]).norm().item()
        v_base_norm = base_v[start:end, :].norm().item()
        
        layer_data["v_heads"].append({
            "head_idx": head_idx,
            "delta_norm": v_delta,
            "base_norm": v_base_norm,
            "relative_change": v_delta / (v_base_norm + 1e-8)
        })
    
    # Analyze O projection per head
    for head_idx in range(config.num_attention_heads):
        start = head_idx * config.head_dim
        end = (head_idx + 1) * config.head_dim
        
        o_delta = (ft_o[:, start:end] - base_o[:, start:end]).norm().item()
        o_base_norm = base_o[:, start:end].norm().item()
        
        layer_data["o_heads"].append({
            "head_idx": head_idx,
            "delta_norm": o_delta,
            "base_norm": o_base_norm,
            "relative_change": o_delta / (o_base_norm + 1e-8)
        })
    
    attention_head_data[f"layer_{layer_idx}"] = layer_data

# Print attention summary
print(f"\n{'Layer':<6} | {'Q Heads (0-7) Relative Change %':^50} | {'Max Head':<10}")
print("-" * 75)

top_attention_heads = []
for layer_idx in range(config.num_hidden_layers):
    layer_data = attention_head_data[f"layer_{layer_idx}"]
    q_changes = [h["relative_change"]*100 for h in layer_data["q_heads"]]
    
    # Find max head
    max_idx = np.argmax(q_changes)
    max_val = q_changes[max_idx]
    
    top_attention_heads.append({
        "layer": layer_idx,
        "head": max_idx,
        "change": max_val
    })
    
    # Visual bar for each head
    bars = ""
    for c in q_changes:
        if c > 6: bars += "█"
        elif c > 5: bars += "▓"
        elif c > 4: bars += "▒"
        else: bars += "░"
    
    print(f"L{layer_idx:<4} | {bars} {' '.join([f'{c:.1f}' for c in q_changes]):^40} | Head {max_idx} ({max_val:.1f}%)")

# 4. PRECISE MLP NEURON ANALYSIS
print("\n" + "=" * 70)
print(f"4️⃣ MLP NEURON ANALYSIS ({config.intermediate_size} NEURONS × 26 LAYERS)")
print("=" * 70)

mlp_neuron_data = {}
all_neuron_changes = []

for layer_idx in range(config.num_hidden_layers):
    # Get all MLP weights
    ft_gate = finetuned_model.model.layers[layer_idx].mlp.gate_proj.weight.data.cpu().float()
    base_gate = base_model.model.layers[layer_idx].mlp.gate_proj.weight.data.cpu().float()
    
    ft_up = finetuned_model.model.layers[layer_idx].mlp.up_proj.weight.data.cpu().float()
    base_up = base_model.model.layers[layer_idx].mlp.up_proj.weight.data.cpu().float()
    
    ft_down = finetuned_model.model.layers[layer_idx].mlp.down_proj.weight.data.cpu().float()
    base_down = base_model.model.layers[layer_idx].mlp.down_proj.weight.data.cpu().float()
    
    # Per-neuron analysis (neuron = column in gate/up, row in down)
    # Combined change: gate + up (input side) + down (output side)
    gate_deltas = (ft_gate - base_gate).norm(dim=1)  # [intermediate_size]
    up_deltas = (ft_up - base_up).norm(dim=1)        # [intermediate_size]
    down_deltas = (ft_down - base_down).norm(dim=0)  # [intermediate_size]
    
    # Total neuron change
    total_deltas = gate_deltas + up_deltas + down_deltas
    
    # Neuron importance (weight magnitude)
    gate_norms = ft_gate.norm(dim=1)
    down_norms = ft_down.norm(dim=0)
    
    # Statistics
    top_k = 20
    top_neurons = total_deltas.topk(top_k)
    
    layer_neuron_data = {
        "layer": layer_idx,
        "total_neurons": config.intermediate_size,
        "mean_delta": total_deltas.mean().item(),
        "std_delta": total_deltas.std().item(),
        "max_delta": total_deltas.max().item(),
        "min_delta": total_deltas.min().item(),
        "top_20_neurons": [
            {"neuron_idx": idx.item(), "delta": val.item()}
            for idx, val in zip(top_neurons.indices, top_neurons.values)
        ],
        # Near-zero analysis (potential dead neurons)
        "near_zero_gate": (gate_norms < 0.01).sum().item(),
        "near_zero_down": (down_norms < 0.01).sum().item(),
    }
    
    mlp_neuron_data[f"layer_{layer_idx}"] = layer_neuron_data
    
    # Track all neurons globally
    for neuron_idx in range(config.intermediate_size):
        all_neuron_changes.append({
            "layer": layer_idx,
            "neuron": neuron_idx,
            "delta": total_deltas[neuron_idx].item(),
            "gate_delta": gate_deltas[neuron_idx].item(),
            "up_delta": up_deltas[neuron_idx].item(),
            "down_delta": down_deltas[neuron_idx].item(),
        })

# Print MLP summary
print(f"\n{'Layer':<6} | {'Mean Δ':>8} | {'Max Δ':>8} | {'Top Neuron':>12} | {'Near-Zero':>10}")
print("-" * 60)

for layer_idx in range(config.num_hidden_layers):
    data = mlp_neuron_data[f"layer_{layer_idx}"]
    top_neuron = data["top_20_neurons"][0]
    print(f"L{layer_idx:<4} | {data['mean_delta']:>8.4f} | {data['max_delta']:>8.4f} | "
          f"N{top_neuron['neuron_idx']:>5} ({top_neuron['delta']:.2f}) | {data['near_zero_gate']:>10}")

# 5. TOP 50 MOST CHANGED NEURONS GLOBALLY
print("\n" + "=" * 70)
print("5️⃣ TOP 50 MOST CHANGED MLP NEURONS (GLOBAL)")
print("=" * 70)

all_neuron_changes.sort(key=lambda x: x["delta"], reverse=True)
top_50_neurons = all_neuron_changes[:50]

print(f"\n{'Rank':<5} | {'Layer':<6} | {'Neuron':<8} | {'Total Δ':>10} | {'Gate Δ':>8} | {'Up Δ':>8} | {'Down Δ':>8}")
print("-" * 75)

for i, n in enumerate(top_50_neurons):
    print(f"{i+1:<5} | L{n['layer']:<5} | N{n['neuron']:<7} | {n['delta']:>10.4f} | "
          f"{n['gate_delta']:>8.4f} | {n['up_delta']:>8.4f} | {n['down_delta']:>8.4f}")

# 6. TOP ATTENTION HEADS (sorted)
print("\n" + "=" * 70)
print("6️⃣ TOP 20 MOST CHANGED ATTENTION HEADS")
print("=" * 70)

top_attention_heads.sort(key=lambda x: x["change"], reverse=True)
print(f"\n{'Rank':<5} | {'Layer':<6} | {'Head':<6} | {'Change %':>10}")
print("-" * 35)
for i, h in enumerate(top_attention_heads[:20]):
    print(f"{i+1:<5} | L{h['layer']:<5} | H{h['head']:<5} | {h['change']:>10.2f}%")

# 7. SAVE ALL DATA
print("\n" + "=" * 70)
print("7️⃣ SAVING PRECISE DATA FOR CIRCUIT MAPPING")
print("=" * 70)

# Save architecture
with open("/workspace/results/weights/architecture.json", "w") as f:
    json.dump(architecture, f, indent=2)

# Save full attention head data
with open("/workspace/results/weights/attention_heads_full.json", "w") as f:
    json.dump(attention_head_data, f, indent=2)

# Save full MLP neuron data
with open("/workspace/results/weights/mlp_neurons_full.json", "w") as f:
    json.dump(mlp_neuron_data, f, indent=2)

# Save top 50 neurons
with open("/workspace/results/weights/top_50_neurons.json", "w") as f:
    json.dump(top_50_neurons, f, indent=2)

# Save top attention heads
with open("/workspace/results/weights/top_attention_heads.json", "w") as f:
    json.dump(top_attention_heads, f, indent=2)

# Create circuit mapping reference
circuit_reference = {
    "architecture": architecture,
    "attention": {
        "total_q_heads": config.num_attention_heads * config.num_hidden_layers,
        "total_kv_heads": config.num_key_value_heads * config.num_hidden_layers,
        "heads_per_layer_q": config.num_attention_heads,
        "heads_per_layer_kv": config.num_key_value_heads,
        "head_dim": config.head_dim,
        "top_20_changed_heads": top_attention_heads[:20]
    },
    "mlp": {
        "total_neurons": config.intermediate_size * config.num_hidden_layers,
        "neurons_per_layer": config.intermediate_size,
        "top_50_changed_neurons": top_50_neurons
    }
}
with open("/workspace/results/weights/circuit_reference.json", "w") as f:
    json.dump(circuit_reference, f, indent=2)

print("   ✅ /workspace/results/weights/architecture.json")
print("   ✅ /workspace/results/weights/attention_heads_full.json")
print("   ✅ /workspace/results/weights/mlp_neurons_full.json")
print("   ✅ /workspace/results/weights/top_50_neurons.json")
print("   ✅ /workspace/results/weights/top_attention_heads.json")
print("   ✅ /workspace/results/weights/circuit_reference.json")

# Cleanup
del base_model
torch.cuda.empty_cache()

print("\n" + "=" * 70)
print("✅ PRECISE ANALYSIS COMPLETE!")
print("=" * 70)

print("\n📊 SUMMARY FOR CIRCUIT MAPPING:")
print(f"   • 26 layers × 8 Q-heads = 208 attention heads analyzed")
print(f"   • 26 layers × 9216 neurons = 239,616 MLP neurons analyzed")
print(f"   • Top changed head: L{top_attention_heads[0]['layer']} H{top_attention_heads[0]['head']} ({top_attention_heads[0]['change']:.2f}%)")
print(f"   • Top changed neuron: L{top_50_neurons[0]['layer']} N{top_50_neurons[0]['neuron']} (Δ={top_50_neurons[0]['delta']:.4f})")
```
